```python
from PIL import Image
from pylab import *
import os
from numpy import *
import imtools.pynb
```

```python
# Images folder
filelist = ["mizore.png", "reina.jpg"]
```

```python
## Converting to .jpg
for infile in filelist:
    outfile = os.path.splitext(infile)[0] + ".jpg"
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
            print("Can not convert", infile, "to .jpg!")
```

```python
## Making miniature with Image.thumbnail()
im = Image.open(filelist[1])
im.thumbnail((555, 100))
im
```

```python
## Resize, Convert, Crop, Paste!! 
##... and Rotate!
mizore_resize_tup = (400, 220)
cropped_region_tup = (50, 50, 150, 150)

im = Image.open(filelist[0]) ## Mizore's here!
im = im.resize(mizore_resize_tup).convert('L')

box = (100, 100, 200, 200) ## Scale of crop
im_region = im.crop(box)
im_region = im_region.transpose(Image.ROTATE_180)

im.paste(im_region, cropped_region_tup)
## Returning to the stock image state
im_orig = Image.open(filelist[0]).convert('L')
im_orig = im_orig.resize(mizore_resize_tup)
im_region_orig = im_orig.crop(cropped_region_tup)
im.paste(im_region_orig, cropped_region_tup)

im.rotate(30)
```
